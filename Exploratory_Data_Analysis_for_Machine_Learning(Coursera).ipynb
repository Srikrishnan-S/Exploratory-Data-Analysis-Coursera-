#### ESTIMATION and INFERENCE :

- **Estimation** : Any numerous procedures used to caculate the value of some property of a population from observations of a sample drawn from the population.
- **Inference** : THe process of drawing conclusions about an underlying population based on a sample or subset of the data.

#### ESTIMATION vs INFERENCE :


- **Estimation**: is the application of an algorithm, for example taking an average:
-> _X_ = **sum**.Xi/n
- It is going to give us the estimation of certain parameter example : **mean** from our sample data.
- **Inference**: involves putting an accuracy on the estimate(e.g. standard error of an average):
- We are trying to understand the underlying distribution of the population.
-> E = **sum**.(((Xi-_X_)^2)/(n-1))^1/2).


#### MACHINE LEARNING AND STATISTICAL INFERENCE :

- **Machine Learning** and **Statistical Inference** are similar.
- In both the cases we are using **data** to learn qualities of a distribution that generated the data.(often termed as **data-generating-prcess**).
- We may care ether about the whole distrbution or just features(e.g.**mean**).
- Machine Learning applications that focus on **understanding parameters** and individual effects involve more tools from **statistical inference**(some applications are focused only on results).

#### PARAMETRIC VS NON-PARAMETRIC :


- The **statistical inference** is about finding the underlying **Data Generating Process** of our data.Then the statistical model is going to be a set of the possible distributions or even regression that data can take.
- **Parametric Model** is a particular type of statistical model. The parametric model constrained to a finite number of parameters, and that rely on some strict assumptions made about the distributions from which that data is pulled.
- **Example of Parametric** : is creating a distribution based on **Normal Distribution**.
- **Non-Parametric Model** mean that the inference will not rely on as many assumptions, such as it will not have to rely on the data being pulled from a particular dstribution, it will be a distribution free inference. it does'nt mean that we know nothing., we will be using insight from the data we have available.
- **Examples of Non-Parametric** : is creating a distrubution based on **Cumilative Distribution**.


#### COSTUMER-LIFETIME-VALUE : 

- **Costumer Lifetime Value** : is an estimate of the costumer's value to the company.
- data related to costumer lifetime value might include : 
-> The expected length of time as a customer.
-> The expected amount spent over time.
-  To estimate lifetime value, we make **assumptions** about the data, namely ow long the costumer is going to last and how much do we think they will spend over time.
- These assumption can be **parametric**(assuming a specific distribution), so assume a **specific distribution**, wether that's **linear over time** or some type of **decrease over time**, or **non-parametric**.

#### PARAMTERIC MODELS : MAXIMUM LIKELIHOOD

- The most common way of estimating parameters in a parametric model is through **maximum likelihood (MLE)**.
- The **Likelyhood Function** is related to probabilty and i a function of the **parameters** of the model.

#### FREQUENTIST AND BAYESIAN STATISTICS : 

- A **Frequentist** is concerned with repeated observations in the limit.
- Processes may be true frequensies, but we're interested in modelling probablities as many, may repeats of an experiment.
-  **Frequentist Approach** : 
-> **Derive** estimates directly from the data with no external influence.
-> **Apply** the **estimates** to the **derived value** to our observed data.

- A **Bayesian** describes parameters by probabilty distributions.
- Before seeing any data, a **prior distribution** is formulated.
- The **prior distribution** is then updated after seeing data.
- After updating, the distribution is called the **posterior distrubution**.

### HYPOTHESIS TESTING

#### HYPOTHESIS TESTING TOPICS : 

- Overview of **Hypothesis Testing**.
- **Bayesian** approach to **Hypothesis Testing**.
- An example of **Hypothesis Testing** involving **coin-tossing**.


#### INTRODUCTION TO HYPOTHESIS TESTING : 

- A **Hypothesis** i a statement about a population parameter.
- We create **Two Hypothesis** :
-> The **Null Hypothesis (H0)**.
-> The **Alternative Hypothesis (H1 or Ha)**.
- We decide which one to call the null depending on how the problem is set up.

#### HYPOTHESIS TESTING : DECISION RULES

- A **Hypothesis Testing Procedure** gives us a rule to decide:
-> For which values of te test statistics do we **accept H0**.
-> For whch values of the test statistics do we **Reject H0** and **Accept H1**.
-  We are using **Hypothesis Testing** in order to decide which two paths to take in the project.

#### HYPOTHESIS TESTING : BAYESIAN APPROACH

- In the **Bayesian Interpretation** we don't get a decision boundary.
- Instead we get updated probabilities of the **Null** and **Alternative Hypothesis**, and see which one is more likely.

#### HYPOTHESIS TESTING EXAMPLES : COIN TOSSING EXAMPLE

- You have **Two Coins** :
-> **Coin 1** has a **70% probability** of coming up **heads**.
-> **Coin 2** has a **50% probability** of coming up **heads**.
- Pick one coin without looking.
- Toss the coin **10 times** and record the number of heads.
- Given the number of heads you see, which of the two coins did you toss?

#### COIN TOSSING EXAMPLE : LIKELIHOOD RATIO

- Given what we know about coins 1 and 2, we can make a table of the probability of seeing x heads out of 10 tosses.
- So **lower numbers** are more likely to be at **0.50 coin**.
- **higher numbers** are more likely to be at **0.70 biased coin**.
- Now we can caculate a **likelihood ratio**, based on the **number of heads** we saw when **tossing** the **unidentified coin**.
- The ratio of which one of our **Null** vs **Alternative Hypothesis** is more likely.

#### HYPOTHESIS TESTING : BAYESIAN INTERPRETATION

- In the **Bayesian Interpretation**, we need **priors** for each **Hypothesis** :
- In this case, we randomly chose the coin to flip
-  Since we have no way, before seeing the data, to determine the coin that was chosen, we just assign 1/2 to each.
- The **prior distributions** are multiplied by the **likelihood ratio**, which does not depend on the priors.
- The **Likelihood Ratio** tells us how we should update the **priors** in reaction to seeing a given set of data.


#### NEYMAN-PEARSON INTERPRETATION : 

- The **Neyman-Pearson Paradigm** is **Non-Bayesian**.
- This gives an **up or down vote** on **H0 vs H1**.

#### TYPE_1 VS TYPE_2 ERROR : 

- **Type_1 Error** : is **incorrectly rejecting** the **Null H0**.
- **Type_2 Error** : is **incorrectly accept** the **Null H0**.

- We will choose, before **testing** the **data**, the levelat whch we will **reject** the **Null Hypothesis**.

#### SIGNIFICANCE LEVEL AND P-VALUES : 

- **SIGNIFICANCE LEVEL**: A **Significance Level (Alpha)** is a probability treshold below which the null hypothesis will be rejected.
- We must choose an **Alpha** before computing the test statistic!
- If we don't, we might be accused of **p-hacking**.
- Choosing **Alpha** is somewhat **arbitary**, but often **p-values** are **.01** or **.05** or even **0.01** if wanted to be **really safe**.

- **P-VALUE**: The probability under the null distribution of result as or more extreme than what was actually observed. A smallest significance level at which the **Null Hypothesis** would be rejected.
- The **Confidence Interval** is the value of the statistic for which we accept the **Null Hypothesis**.
 

#### F-STATISTICS : 

- An **F-Test** is any statistical test in which the test statistics has an **F-Distribution** under the **Null Hypothesis**.
- It is most often used when comparing **Statistical Models** that have been fitted to a **datset**.
